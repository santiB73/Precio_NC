```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import mstats
```

```python
df = pd.read_excel(r"C:\Users\User\Documents\Precio_VN_NC.xlsx")

df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

# Plot a histogram of the 'Valor_Nominal' variable
plt.hist(df['Valor_Nominal'], bins=30, color='blue', alpha=0.7)
plt.xlabel('Valor_Nominal')
plt.ylabel('Frequency')
plt.title('Histogram of Valor_Nominal')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

n, bins, _ = plt.hist(df['Valor_Nominal'], bins=200, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram for Valor_Nominal')
plt.show()

# Calculate the total number of data points
total_data_points = len(df['Valor_Nominal'])

# Calculate the percentage for each bin
for i, bin_value in enumerate(bins[:-1]):
    bin_start = bins[i]
    bin_end = bins[i + 1]
    bin_frequency = n[i]
    bin_percentage = (bin_frequency / total_data_points) * 100
    print(f"Bean {i + 1} ({bin_start:.0f}-{bin_end:.0f}): {bin_percentage:.2f}%")
```

```python
statistics = df['Valor_Nominal'].describe()
statistics_no_sci = statistics.apply(lambda x: '{:.2f}'.format(x))
print(statistics_no_sci)
```
